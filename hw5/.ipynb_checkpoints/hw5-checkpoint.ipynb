Run k-means on the data, for k = 2, k = 5, and k = 20. Try a few (at least 5 each) different initializations and check to see whether they find the same solution; if not, pick the one with the best score. For
the chosen assignment for each k, include a plot with the data, colored by assignment, and the cluster
centers. You can plot the points colored by assignments using ml.plotClassify2D(None,X,z),
where z are the resulting cluster assignments of the data. You will have to additionally plot the centers
yourself. 

```python
K = [2, 5, 20]
for k in K:
    km_z, km_c, km_sumd = ml.cluster.kmeans(iris2, k)
    plt.figure()
    ml.plotClassify2D(None, iris2, km_z)   
    plt.scatter(*zip(*km_c), c = "red", marker = "x")
```

Run agglomerative clustering on the data, using single linkage and then again using complete linkage,
each with 2, 5, and then 20 clusters (using ml.cluster.agglomerative from cluster.py).
Again, plot with color the final assignment of the clusters. (This algorithm has no initialization issues;
so you do not have to try multiple initializations.)

```python
#run with method = min, method = max for single & complete linkage 
K = [2, 5, 20]
mthd = ["min", "max"]
for m in mthd:
    for k in K:
        km_z, km_join = ml.cluster.agglomerative(iris2, k, method = m)
        plt.figure()
        title = str(m) + ", " + str(k)
        plt.title(title)
        ml.plotClassify2D(None, iris2, km_z)   
    #plt.scatter(*zip(*km_c), c = "red", marker = "x")
```

#Single linkage: we merge the clusters whose two closest members have minimum distance. 
#Complete linkage: distance between clusters is the furthest distance among members. 

4) Describe similarities and differences in the results from the agglomerative clustering and k-means.

The results from k-means are generally more intuitive, and I think probably better. With single linkage agglomerative clustering, we end up with clusters that are pointlessly small, like the yellow dot on k = 5. Between single and complete linkage, qualitatively complete linkage seems to have better results, that are more similar to k-means. Complete linkage is also more sensitive to the empty space inbetween clusters, while in single linkage you can get weird results that seem to ignore the spaces.

Subtract the mean of the face images (X0 = X − µ) to make your data zero-mean. (The mean should
be of the same dimension as a face, 576 pixels.) Plot the mean face. (5 points)

```python
mu= np.mean(X, axis=0, keepdims=True)
X0 = X - mu
print(mu.shape)
print(X0.shape)
```

```python
img = np.reshape(mu, (24, 24))
plt.imshow(img.T, cmap = "gray")
```

Use scipy.linalg.svd to take the SVD of the data so that 

X0 = U · diag(S) · Vh

Since the number of data is larger than the number of dimensions, there are at most 576 non-zero
singular values; you can use full matrices=False to avoid using a lot of memory. As in the
slides, we suggest computing W = U.dot( np.diag(S) ) so that X0 ≈ W · Vh. Print the
shapes of W and Vh. (10 points)

```python
import scipy.linalg
```

```python
U, S, Vh = scipy.linalg.svd(X0, False)
W = U.dot(np.diag(S))
```

```python
print(W.shape)
```

```python
print(Vh.shape)
```

Methods like PCA are often called “latent space” methods, as the coefficients can be interpreted as
a new geometric space in which the data are being described. To visualize this, choose a few faces
(say 25), and display them as images in a two-dimensional plane (as in the lecture), where the 
twodimensional coordinates are given by their coefficients on the first two principal components:

```python
idx = np.random.randint(low = 0, high = 4915, size = 20) 
#idx = ... # pick some data (randomly or otherwise); an array of integerindices
import mltools.transforms
coord,params = ml.transforms.rescale(W[:,0:2]) # normalize scale of "W"locations
plt.figure(figsize=(8, 8)); 
#plt.hold(True); # you may need this for pyplot
for i in idx:
# compute where to place image (scaled W values) & size
    loc = (coord[i,0],coord[i,0]+0.5, coord[i,1],coord[i,1]+0.5)
    img = np.reshape( X[i,:], (24,24) ) # reshape to square
    plt.imshow( img.T , cmap="gray", extent=loc ) # draw each image
    plt.axis( (-2,2,-2,2) )
```
